```python
# Import dependencies
import pymongo
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':20}

# Use count_documents to display the number of documents in the result
q1 = establishments.count_documents(query)
print("docs with hygiene = 20: ",q1,'\n')

# Display the first document in the results using pprint
q1_results = establishments.find(query)
pprint(q1_results[0])
```

```python
# Convert the result to a Pandas DataFrame
q1_df = pd.json_normalize(q1_results)
# Display the number of rows in the DataFrame
q1_df.shape
# Display the first 10 rows of the DataFrame
q1_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'RatingValue':{'$lte':4}}

# Use count_documents to display the number of documents in the result
print("docs with ratings <= 4: ",establishments.count_documents(query),'\n')
# Display the first document in the results using pprint
q2_results = establishments.find(query)
pprint(q2_results[0])
```

```python
# Convert the result to a Pandas DataFrame
q2_df = pd.json_normalize(q2_results)
# Display the number of rows in the DataFrame
q2_df.shape
# Display the first 10 rows of the DataFrame
q2_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
establishments.create_index([('geocode',pymongo.GEO2D)],name='geocode')
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = {'$match': {'geocode.latitude':{'$gte': 51.49014200-degree_search,'$lte':51.49014200+degree_search}}}
longitude = {'$match': {'geocode.longitude':{'$gte': 0.08384000-degree_search,'$lte':0.08384000+degree_search}}}

query = {'$match': {'RatingValue': 4}}
sort =  {'$sort':{'scores.Hygeine':1}}
pipeline = [query, latitude, longitude,sort]

# Print the results
q3_results = list(establishments.aggregate(pipeline))
print(q3_results)
```

```python
# Convert result to Pandas DataFrame
q3_df = pd.json_normalize(q3_results)
q3_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
q4_match = {'$match':{'scores.Hygiene':0}}
q4_group = {'$group':{'_id':'$LocalAuthorityName','count': {'$sum': 1}}}
q4_sort = {'$sort': { 'count': -1}}
q4_pipeline = [q4_match,q4_group,q4_sort]
q4_results = list(establishments.aggregate(q4_pipeline))
# Print the number of documents in the result
print('docs = ',len(q4_results))
# Print the first 10 results
print('first 10: \n',q4_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
q4_df = pd.DataFrame(q4_results)
# Display the number of rows in the DataFrame
q4_df.shape
# Display the first 10 rows of the DataFrame
q4_df.head(10)
```
